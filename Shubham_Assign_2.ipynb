Averge filtering with kernel size 3x3 is as follows :

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image
image = cv2.imread('image_1.jpg')

# I am converting the image to grayscale image
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# I am creating a zero-filled numpy array for the filtered image
filtered_image = np.zeros_like(gray)

# Apply the 3x3 average filter to the image
for i in range(1, gray.shape[0]-1):
    for j in range(1, gray.shape[1]-1):
        filtered_image[i,j] = np.mean(gray[i-1:i+2, j-1:j+2])

# Display the filtered image

cv2_imshow( filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# load the image
img = cv2.imread('image_2.jpg')

# I am finding  the dimensions of the image
height, width, channels = img.shape

# As asked in question I am setting the percentage of pixels to add noise 
percentage = 0.1

# calculate the number of pixels to add noise to
num_pixels = int(height * width * percentage)

# set the probability of a pixel being set to white or black
prob = 0.5

# add noise to the pixels
for i in range(num_pixels):
    x = np.random.randint(0, width)
    y = np.random.randint(0, height)
    if np.random.rand() < prob:
        img[y, x] = [0, 0, 0]
    else:
        img[y, x] = [255, 255, 255]

# display the noisy image
cv2_imshow(img)
```

median filtering with 3x3 kernel size is as follows :

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# read the image
img = cv2.imread('image_2.jpg', 0)

# define the kernel size
ksize = 3

# I am creating a padding of zeros around the image
pad_size = ksize // 2
img_padded = cv2.copyMakeBorder(img, pad_size, pad_size, pad_size, pad_size, cv2.BORDER_CONSTANT, value=0)

# To create an empty output image
img_out = np.zeros_like(img)

# iterate over each pixel of the image
for i in range(pad_size, img.shape[0] + pad_size):
    for j in range(pad_size, img.shape[1] + pad_size):
        # get the pixel neighborhood
        neighborhood = img_padded[i - pad_size:i + pad_size + 1, j - pad_size:j + pad_size + 1]
        # apply median filtering on the neighborhood
        median_val = np.median(neighborhood)
        # set the pixel value of the output image to the median value
        img_out[i - pad_size, j - pad_size] = median_val

# display the output image
cv2_imshow(img_out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**Que -3)** **Gaussian filtering with kernel size 3x3 is as follows :** 

```python
import cv2
import numpy as np

# Load input image
img = cv2.imread('image_3.jpg')

# Define Gaussian kernel size and standard deviation
k_size = 3
sigma = 1.5

# Create kernel grid
kernel = np.zeros((k_size, k_size), dtype=np.float32)
for i in range(k_size):
    for j in range(k_size):
        x = i - k_size // 2
        y = j - k_size // 2
        kernel[i, j] = np.exp(-(x**2 + y**2) / (2 * sigma**2))
kernel = kernel / np.sum(kernel)

# Pad input image
pad_size = k_size // 2
padded_img = cv2.copyMakeBorder(img, pad_size, pad_size, pad_size, pad_size, cv2.BORDER_CONSTANT)

# Appling filter as follows
output = np.zeros_like(img)
for i in range(pad_size, padded_img.shape[0]-pad_size):
    for j in range(pad_size, padded_img.shape[1]-pad_size):
        region = padded_img[i-pad_size:i+pad_size+1, j-pad_size:j+pad_size+1]
        region_gray = cv2.cvtColor(region, cv2.COLOR_BGR2GRAY)
        output[i-pad_size, j-pad_size] = np.sum(region_gray * kernel)

# Display the filtered image
cv2_imshow( output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Que-4) 3-level Gaussian and Laplacian Pyramid


```python
import cv2
from google.colab.patches import cv2_imshow


# Load an image
img = cv2.imread('image_3.jpg')

# Define kernel size for Gaussian Pyramid
gauss_kernel_size = (5,5 )

# Define kernel size for Laplacian Pyramid
lap_kernel_size = (3, 3)

# Generate 3-level Gaussian Pyramid
gauss_pyr = [img]
for i in range(3):
    gauss_pyr.append(cv2.pyrDown(gauss_pyr[i], gauss_kernel_size))

# Generate 3-level Laplacian Pyramid
lap_pyr = [gauss_pyr[2]]
for i in range(2, -1, -1):
    lap = cv2.subtract(gauss_pyr[i], cv2.pyrUp(gauss_pyr[i+1], lap_kernel_size))
    lap_pyr.append(lap)

# Display the pyramids
for i in range(4):
    cv2_imshow( gauss_pyr[i])
for i in range(3):
    cv2_imshow(lap_pyr[i])
cv2.waitKey(0)
cv2.destroyAllWindows()


```

```python
import cv2
import numpy as np
import pywt

# Load the input image
img = cv2.imread('image_3.jpg', 0)


SP = 0.05  # Set the noise ratio for adding salt and pepper noise
noise = np.zeros(img.shape, np.uint8)
cv2.randu(noise, 0, 255)

img_noise = img.copy()
img_noise[noise < SP * 255] = 0
img_noise[noise > (1 - SP) * 255] = 255

# To perform laplacian filtering on the noisy image
laplacian_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
img_laplacian = cv2.filter2D(img_noise, -1, laplacian_kernel)

# To perform the final operation to get the output image
img_output = cv2.add(img, img_laplacian)

# Perform wavelet decomposition using HAAR filter
coeffs = pywt.wavedec2(img_output, 'haar', level=2)


# Set the approximation coefficients to zero to remove high frequency components
coeffs[0] *= 0

# Reconstruct the image from the modified wavelet coefficients
img_reconstructed = pywt.waverec2(coeffs, 'haar')

# Display the resultant image

cv2_imshow(img_reconstructed)
```
